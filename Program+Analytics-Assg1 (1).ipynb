```python
import pandas as pd
```

```python
#Function to skil comments
def skip_comments(file):
    for line in file:
        if not line.strip().startswith('#'):
            yield line
```

```python
#Program to load file data into dataframe
data = []
f = open("/Users/rujutadawhkar/Downloads/co2_alaska.txt", "r")
for line in skip_comments(f):
    import re
    t = re.split(r'[;,\s]\s*', line)
    data.append(t)
df = pd.DataFrame(data)
df
```

```python
# For 2nd file
data2 = []
f1 = open("/Users/rujutadawhkar/Downloads/co2_hawaii.txt", "r")
for line in skip_comments(f1):
    import re
    t1 = re.split(r'[;,\s]\s*', line)
    data2.append(t1)
df1 = pd.DataFrame(data2)
df1
```

```python
#check datatype of frame
df.dtypes
```

```python
#converting datatype of 7th column (Value) to float
df[7] = df[7].astype(float)
df1[7] = df1[7].astype(float)
```

```python
#filter valid data such that it had "..." value in 16th (qc flag) column 
filterflag = df[df[16] == "..."]
filterflag1 = df1[df1[16] == "..."]
```

```python
filterflag1
```

```python
#Find max value of data year wise in hawaii and alaska
max1_alaska = filterflag.groupby([1], sort=False)[7].max()
max_hawaii = filterflag1.groupby([1], sort=False)[7].max()
```

```python
# Convert max data year wise into dataframe
max2_alaska = max1_alaska.add_suffix('').reset_index()
max2_alaska = max2_alaska.rename(columns={1: 'Year'})
max2_alaska = max2_alaska.rename(columns={7: 'Maximum Value'})
max2_alaska
```

```python
# Convert max data year wise into dataframe
max_hawaii1 = max_hawaii.add_suffix('').reset_index()
max_hawaii1 = max_hawaii1.rename(columns={1: 'Year'})
max_hawaii1 = max_hawaii1.rename(columns={7: 'Maximum Value'})
max_hawaii1
```

```python
# Filter min value data year wise
mean1_alaska = filterflag[7].groupby([df[1]]).mean()
mean1_alaska
mean_hawaii = filterflag1[7].groupby([df1[1]]).mean()
mean_hawaii
```

```python
# Convert mean data year wise into dataframe
mean2_alaska = mean1_alaska.add_suffix('').reset_index()
mean2_alaska = mean2_alaska.rename(columns={1: 'Year'})
mean2_alaska = mean2_alaska.rename(columns={7: 'Mean Value'})
mean2_alaska
```

```python
# Convert max data year wise into dataframe
mean_hawaii1 = mean_hawaii.add_suffix('').reset_index()
mean_hawaii1 = mean_hawaii1.rename(columns={1: 'Year'})
mean_hawaii1 = mean_hawaii1.rename(columns={7: 'Mean Value'})
mean_hawaii1
```

```python
# Merge dataframe with MAx and Min data yearwise
final_alaska = pd.merge(max2_alaska,mean2_alaska, on='Year')
final_alaska
```

```python
# Merge max and mean data year wise
final_hawaii = pd.merge(max_hawaii1,mean_hawaii1, on='Year')
final_hawaii
```

```python
# Drop Year column such that only Mean value is left for percentage change
mean3_alaska = mean2_alaska.drop('Year', 1)
mean3_alaska
mean_hawaii2 = mean_hawaii1.drop('Year', 1)
mean_hawaii2
```

```python
# Convert this data frame to a list to us pct.change() function
my_list_alaska = mean3_alaska["Mean Value"].tolist()
my_list_alaska
my_list_hawaii = mean_hawaii2["Mean Value"].tolist()
my_list_hawaii
```

```python
# Find percent mean chanage using pct.change function
alaska_list = pd.Series(my_list_alaska)
Percent_Change_alaska = alaska_list.pct_change()
Percent_Change_alaska
hawaii_list = pd.Series(my_list_hawaii)
Percent_Change_hawaii = hawaii_list.pct_change()
Percent_Change_hawaii
```

```python
# Convert it to data frame
Percent_alaska = pd.DataFrame({'PercentChange':Percent_Change_alaska})
Percent_alaska
Percent_hawaii = pd.DataFrame({'PercentChange':Percent_Change_hawaii})
Percent_hawaii
```

```python
# Concatenate all dataframe to add PErcent Change Column
final1_alaska = pd.concat([final_alaska, Percent_alaska], axis=1)
final1_alaska['PercentChange'] = final1_alaska['PercentChange']*100
final1_alaska
```

```python
# Concatenate all dataframe to add PErcent Change Column
final_hawaii1 = pd.concat([final_hawaii, Percent_hawaii], axis=1)
final_hawaii1['PercentChange'] = final_hawaii1['PercentChange']*100
final_hawaii1
```

```python
# add Respective country value column to dataframes
final1_alaska['Country'] = 'Alaska'
final1_alaska
final_hawaii1['Country'] = 'Hawaai'
final_hawaii1
```

```python
#Concatename hawaii and alaska dataframes
df_new = pd.concat([final1_alaska, final_hawaii1])
df_new
```

```python
# Create a pivot table and round decimal to 2 places
d2 = df_new.pivot_table(index=['Year'],columns='Country', values=('Maximum Value','Mean Value','PercentChange'),fill_value=0)
d3 = d2.round(2)
d3
```

```python
#Write data to txt file
d3.to_csv(r'/Users/rujutadawhkar/Downloads/output1.txt',sep=' ')
```
